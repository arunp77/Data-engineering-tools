```python
from decouple import config

SPOTIPY_CLIENT_ID = config('SPOTIPY_CLIENT_ID')
SPOTIPY_CLIENT_SECRET = config('SPOTIPY_CLIENT_SECRET')
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

client_id = SPOTIPY_CLIENT_ID
client_secret = SPOTIPY_CLIENT_SECRET

client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
playlist_id = '44kgWZRt5jmmgppH0qGszg'  # Top Songs playlist ID
country = 'US'  # Replace 'US' with the desired country code (in quotation marks)
results = sp.playlist_tracks(playlist_id, market=country)
```

```python
results
```

```python
for idx, track in enumerate(results['items'], start=1):
    song_name = track['track']['name']
    artist = track['track']['artists'][0]['name']
    album = track['track']['album']['name']
    print(f"{idx}. {song_name} by {artist} from {album}")
```
