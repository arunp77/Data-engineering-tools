```python
import pandas as pd
import numpy as np
import random
import getindianname as name  # Replace with the correct library name

# Generate a list of male names
num_male_names = 600
male_names = [name.male() for _ in range(num_male_names)]

# Generate a list of female names
num_female_names = 300
female_names = [name.female() for _ in range(num_female_names)]

# Generate a list of random names
num_random_names = 100
random_names = [name.randname() for _ in range(num_random_names)]

# Combine the lists into a single list
all_names = male_names + female_names + random_names

# Create a DataFrame for the first set of names and genders
df_names = pd.DataFrame({'Employee_Name': all_names, 'Gender': ['Male'] * num_male_names + ['Female'] * num_female_names + ['Random'] * num_random_names})

# Set seed for reproducibility
np.random.seed(42)

# Generate a synthetic dataset with 1000 rows
num_rows = 1000

# Demographic data
gender = np.random.choice(['Male', 'Female'], size=num_rows)
ethnicity = np.random.choice(['Hindu', 'Muslim', 'Christian', 'Sikh', 'Other'], size=num_rows)
caste = np.random.choice(['General', 'OBC', 'SC', 'ST', 'Other'], size=num_rows)
region = np.random.choice(['North', 'South', 'East', 'West', 'Central'], size=num_rows)

# Inclusion initiatives and impact
diversity_training = np.random.choice(['Yes', 'No'], size=num_rows, p=[0.7, 0.3])
employee_support_networks = np.random.choice(['Yes', 'No'], size=num_rows, p=[0.6, 0.4])

# Cultural sensitivity and awareness
cultural_sensitivity_programs = np.random.choice(['Yes', 'No'], size=num_rows, p=[0.8, 0.2])

# Create a DataFrame for the second set of data
data = {
    'Employee_Name': all_names,  # Include the 'Employee_Name' column
    'Gender': gender,
    'Ethnicity': ethnicity,
    'Caste': caste,
    'Region': region,
    'Diversity_Training': diversity_training,
    'Employee_Support_Networks': employee_support_networks,
    'Cultural_Sensitivity_Programs': cultural_sensitivity_programs
}

df_additional = pd.DataFrame(data)

# Generate 'Role' column
roles = ['Manager', 'Supervisor', 'Team Lead', 'Specialist', 'Analyst', 'Coordinator', 'Associate']
df_additional['Role'] = np.random.choice(roles, size=num_rows)

# Generate 'Department' column
departments = ['HR', 'Finance', 'IT', 'Marketing', 'Operations']
df_additional['Department'] = np.random.choice(departments, size=num_rows)
```

# Some important variable

1. **'Diversity_Training':** Indicates whether an employee has undergone diversity training. Values can be 'Yes' or 'No'.
2. '**Employee_Support_Networks**': Indicates whether an employee is part of any support networks within the organization. Values can be 'Yes' or 'No'.
3. '**Cultural_Sensitivity_Programs**': Indicates whether an employee has participated in cultural sensitivity programs. Values can be 'Yes' or 'No'.
4. '**Role**': Represents the job role or position of the employee within the organization.
5. '**Department**': Represents the department or division to which the employee belongs.
6. '**Overall_Satisfaction**': Represents the overall satisfaction level of the employee, likely based on a survey or assessment.
7. '**Diversity_Satisfaction**': Represents the satisfaction level of the employee regarding diversity initiatives within the organization.
8. '**Inclusion_Satisfaction**': Represents the satisfaction level of the employee regarding inclusion initiatives within the organization.
9. '**Accessibility_Programs**': Indicates whether an employee has participated in accessibility programs. Values can be 'Yes' or 'No'.
10. '**Collaboration_Score**': Represents a score indicating the level of collaboration the employee engages in.
11. '**Equal_Opportunity_Policy**': Indicates adherence to the equal opportunity policy. Values can be 'Yes' or 'No'.
12. '**External_Partnerships**': Indicates whether an employee has participated in external partnerships. Values can be 'Yes' or 'No'.

These columns capture various aspects related to employee training, satisfaction, involvement in programs, and adherence to policies within the organization. The specific interpretation may depend on the context and how the data was collected.
